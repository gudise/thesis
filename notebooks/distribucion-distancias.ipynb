En este cuaderno estudio cómo se distribuyen las distancias entre puntos, en comparación con cómo se distribuyen los puntos en sí mismos. Este problema surge al tratar de ajustar un set de distancias de Hamming a una cierta distribución de probabilidad: en un conjunto de 100 puntos, típicamente construimos los sets de respuestas haciendo todas las comparaciones posibles, i.e., 100*99/2 puntos, y estudiamos cómo se distribuyen estas distancias. Sin embargo, a la hora de ajustarlos a una curva, típicamente la bondad de ese ajuste depende del tamaño del set experimental, y aquí es donde está la duda: ¿cuál es el tamaño de ese set? ¿100*99/2? Ciertamente ese es el tamaño del conjunto de puntos que estamos ajustando, pero hay algo raro en estimar la bondad del ajuste a curvas de simulación generadas con 100*99/2 grados de libertad cuando *de hecho*, únicamente estamos generando 00 valores aleatorio.

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
## Generamos 100 pts flotantes distribuidos como norm
set_original = st.norm.rvs(size=100)
set_distancias = [set_original[i]-set_original[j] for i in range(100) for j in range(i+1,100,1)]
```

```python
plt.hist(set_distancias, rwidth=0.9)
plt.show()
```
