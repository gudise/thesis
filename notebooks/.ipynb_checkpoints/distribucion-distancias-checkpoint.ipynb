```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = range(101)
set_binom = list(st.binom.rvs(n=100,p=0.5,size=10000))
set_poisson = list(st.poisson.rvs(mu=10,size=10000))

plt.hist(set_binom, rwidth=0.9, bins=100, range=(0,100), density=True)
plt.show()

plt.hist(set_poisson, rwidth=0.9, bins=100, range=(0,100), density=True)
plt.show()

plt.hist(set_binom+set_poisson, rwidth=0.9, bins=100, range=(0,100), density=True)
plt.show()
```
